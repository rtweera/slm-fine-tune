```python
y_pred = clf.predict(embed_test_X)
```

```python
accuracy = accuracy_score(test_y, y_pred)
recall = recall_score(test_y, y_pred, average='weighted')
precision = precision_score(test_y, y_pred, average='weighted')
f1 = f1_score(test_y, y_pred, average='weighted')
cm = confusion_matrix(test_y, y_pred)
print(classification_report(test_y, y_pred, target_names=['Negative', 'Neural', 'Positive']))
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
